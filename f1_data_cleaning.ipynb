```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading all datasets

circuits = pd.read_csv('Dataset/circuits.csv')
constructor_results = pd.read_csv('Dataset/constructor_results.csv')
constructor_standings = pd.read_csv('Dataset/constructor_standings.csv')
constructors = pd.read_csv('Dataset/constructors.csv')
driver_standings = pd.read_csv('Dataset/driver_standings.csv')
driver_standings = pd.read_csv('Dataset/driver_standings.csv')
drivers = pd.read_csv('Dataset/drivers.csv')
lap_times = pd.read_csv('Dataset/lap_times.csv')
pit_stops = pd.read_csv('Dataset/pit_stops.csv')
qualifying = pd.read_csv('Dataset/qualifying.csv')
races = pd.read_csv('Dataset/races.csv')
results = pd.read_csv('Dataset/results.csv')
seasons = pd.read_csv('Dataset/seasons.csv')
sprint_results = pd.read_csv('Dataset/sprint_results.csv')
status = pd.read_csv('Dataset/status.csv')
```

```python
# Checking is duplicate data exists

print(circuits.duplicated().sum())
print(constructor_results.duplicated().sum())
print(constructor_standings.duplicated().sum())
print(constructors.duplicated().sum())
print(driver_standings.duplicated().sum())
print(driver_standings.duplicated().sum())
print(drivers.duplicated().sum())
print(lap_times.duplicated().sum())
print(pit_stops.duplicated().sum())
print(qualifying.duplicated().sum())
print(races.duplicated().sum())
print(results.duplicated().sum())
print(seasons.duplicated().sum())
print(sprint_results.duplicated().sum())
print(status.duplicated().sum())
```

```python
# Checking for null (NaN) values

print(circuits.isnull().sum())
print(constructor_results.isnull().sum())
print(constructor_standings.isnull().sum())
print(constructors.isnull().sum())
print(driver_standings.isnull().sum())
print(driver_standings.isnull().sum())
print(drivers.isnull().sum())
print(lap_times.isnull().sum())
print(pit_stops.isnull().sum())
print(qualifying.isnull().sum())            # Only this contains NaN values 
print(races.isnull().sum())
print(results.isnull().sum())
print(seasons.isnull().sum())
print(sprint_results.isnull().sum())
print(status.isnull().sum())
```

```python
# 1. circuits.csv
# There's no need of 'lat,lng,alt,url' 


circuits_new = circuits.copy()
circuits_new.drop(columns=['lat', 'lng', 'alt', 'url'], inplace=True)
```

```python
# 2. constructor_results.csv
# There's '\N' in entire of 'Status' column
# There's no need of it


constructor_results_new = constructor_results.copy()
constructor_results_new.drop(columns=['status'], inplace=True)
```

```python
# 3. constructor_standings.csv
# Nothing to change here!


constructor_standings_new = constructor_standings.copy()
```

```python
# 4. constructors.csv
# No need of 'url'


constructors_new = constructors.copy()
constructors_new.drop(columns=['url'], inplace=True)
```

```python
# 5. driver_standings.csv
# Nothing to change here!


driver_standings_new = driver_standings.copy()
```

```python
# 6. drivers.csv
# Drop 'url' and replace '\N' with 'NA'


drivers_new = drivers.copy()
drivers_new['number'] = drivers_new['number'].replace('\\N', 'NA')
drivers_new.drop(columns=['url'], inplace=True)
```

```python
# 7. lap_times.csv
# Nothing to change here!
lap_times_new = lap_times.copy()

# 8. pit_stops.csv
# Nothing to change here!
pit_stops_new = pit_stops.copy()
```

```python
# 9. qualifying.csv
# Handling null values


qualifying_new = qualifying.copy()

qualifying_new['q2'] = qualifying_new['q2'].fillna("NA")
qualifying_new['q3'] = qualifying_new['q3'].fillna("NA")

qualifying_new['q1'] = qualifying_new['q1'].replace('\\N', "NA")
qualifying_new['q2'] = qualifying_new['q2'].replace('\\N', "NA")
qualifying_new['q3'] = qualifying_new['q3'].replace('\\N', "NA")
```

```python
# 10. races.csv
# Dropping these columns 'time, url, fp1_time, fp2_time, fp3_time, quali_time, sprint_time'


races_new = races.copy()
races_new.drop(columns=['time', 'url', 'fp1_time', 'fp2_time', 'fp3_time', 'quali_time', 'sprint_time'], inplace=True)
```

```python
# 11. results.csv
# Handling Null values

results_new = results.copy()
results_new.replace('\\N', 'NA', inplace=True)
```

```python
# 12. seasons.csv
# Handling '\N' values


seasons_new = seasons.copy()
seasons_new.replace('\\N', 'NA', inplace=True)
seasons_new.drop(columns='url', inplace=True)
```

```python
# 12. sprint_results.csv
# Handling '\N' values 


sprint_results_new = sprint_results.copy()
sprint_results_new.replace('\\N', 'NA', inplace=True)
```

```python
# status.csv
# Nothing to change here!
status_new = status.copy()
```

```python
# Saving cleaned datasets

circuits_new.to_csv('Cleaned_datasets/circuits_cleaned.csv', index=False)
constructor_results_new.to_csv('Cleaned_datasets/constructor_results_cleaned.csv', index=False)
constructor_standings_new.to_csv('Cleaned_datasets/constructor_standings_cleaned.csv', index=False)
constructors_new.to_csv('Cleaned_datasets/constructors_cleaned.csv', index=False)
driver_standings_new.to_csv('Cleaned_datasets/driver_standings_cleaned.csv', index=False)
drivers_new.to_csv('Cleaned_datasets/drivers_cleaned.csv', index=False)
lap_times_new.to_csv('Cleaned_datasets/lap_times_cleaned.csv', index=False)
pit_stops_new.to_csv('Cleaned_datasets/pit_stops_cleaned.csv', index=False)
qualifying_new.to_csv('Cleaned_datasets/qualifying_cleaned.csv', index=False)
races_new.to_csv('Cleaned_datasets/races_cleaned.csv', index=False)
results_new.to_csv('Cleaned_datasets/results_cleaned.csv', index=False)
seasons_new.to_csv('Cleaned_datasets/seasons_cleaned.csv', index=False)
sprint_results_new.to_csv('Cleaned_datasets/sprint_results_cleaned.csv', index=False)
status_new.to_csv('Cleaned_datasets/status_cleaned.csv', index=False)
```
